```python
import numpy as np
import matplotlib.pyplot as plt
from doggos.fuzzy_sets import Type1FuzzySet, IntervalType2FuzzySet
from doggos.utils.membership_functions import gaussian, sigmoid, trapezoidal
```

# Plot type 1 fuzzy sets

```python
t1_set1 = Type1FuzzySet(sigmoid(0, 1))
t1_set2 = Type1FuzzySet(sigmoid(-1, 2))
t1_set3 = Type1FuzzySet(sigmoid(-1, -2))
t1_set4 = Type1FuzzySet(gaussian(0, 1))
t1_set5 = Type1FuzzySet(gaussian(0, 4))
```

```python
t1_set1.plot(domain=np.arange(-10, 10, 0.01), title='Sigmoid-mf fuzzy set', label=['mf'])
plt.show()
```

```python
domain = np.arange(-5, 5, 0.01)
y_range = (0, 1.1)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

fig.suptitle('Two sigmoid-mf fuzzy sets')
t1_set2.plot(ax[0], label='sigmoid(-1, 2)', domain=domain, y_range=y_range, color='red')
t1_set3.plot(ax[1], label='sigmoid(-1, -2)', domain=domain, y_range=y_range, color=['pink'])
plt.show()
```

```python
domain = np.arange(-10, 10, 0.01)
y_range = (0, 1.1)
fig, ax = plt.subplots()
fig.suptitle('Gaussian-mf fuzzy sets')
t1_set4.plot(ax, domain=domain, y_range=y_range, color='red', label='N(0, 1)')
t1_set5.plot(ax, domain=domain, y_range=y_range, color=[], label='N(0, 4)')
plt.show()
```

```python
t2_set1.plot(
    domain=np.arange(-10, 10, 0.01),
    y_range=(0, 1.1),
    label=['lower mf', 'upper mf'],
    color=[]
)
```
